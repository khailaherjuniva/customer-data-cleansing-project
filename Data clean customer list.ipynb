```python
import pandas as pd
```

```python
df=pd.read_excel(r"C:\Users\aurora\Documents\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace({
    "Yes": "Y",
    "No": "N"
})

df
```

```python
df = df[df["Do_Not_Contact"] != "Y"]
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace(r'[^0-9]', '', regex=True)
df
```

```python
df = df[df["Phone_Number"] != ""]
df
```

```python
df=df.drop(columns="Not_Useful_Column")
df
```

```python
#df["Last_Name"]=df["Last_Name"].str.lstrip("...")
#df["Last_Name"]=df["Last_Name"].str.lstrip("/")
#df["Last_Name"]=df["Last_Name"].str.lstrip("_")

df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
df["Phone_Number"] = (
    df["Phone_Number"]
    .fillna("")                      
    .astype(str)                     
    .str.replace(r'[^0-9]', '', regex=True)  
    .apply(lambda x: f"{x[:3]}-{x[3:6]}-{x[6:10]}" if len(x) == 10 else "")
)
df
```

```python
df["Paying Customer"] = df["Paying Customer"].replace({
    "Yes": "Y",
    "No": "N"
})

df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace({
    "Yes": "Y",
    "No": "N"
})

df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].fillna("N")
df["Paying Customer"] = df["Paying Customer"].fillna("N")

df
```

```python
df = df[df["Do_Not_Contact"] != "Y"]
df
```

```python
df = df[df["Phone_Number"] != ""]
df
```

```python
df = df.reset_index(drop=True)
df
```
